```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

y = df["Survived"]

features = ['Pclass','Sex','SibSp','Parch']
X = pd.get_dummies(df[features])
X_test = pd.get_dummies(af[features])

X, X_test = X.align(X_test, join='outer', axis=1)

model_rf = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model_lr = LogisticRegression()

model_rf.fit(X, y)
model_lr.fit(X, y)

predictions_rf = model_rf.predict(X_test)
predictions_lr = model_lr.predict(X_test)

output_rf = pd.DataFrame({'PassengerId': af["PassengerId"], 'Survived': predictions_rf})
output_lr = pd.DataFrame({'PassengerId': af["PassengerId"], 'Survived': predictions_lr})

print("Random Forest Accuracy:",  model_rf.score(X, y)*100)
print("Logistic Regression Accuracy:",  model_lr.score(X, y)*100)

print("\n")
print("Thank You!")
```
